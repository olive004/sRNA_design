The above graphs are a summary of how the experimental data relates to the design attributes of the sRNA, based on ChiX as a template.

## Looking at absolute log fold changes
Hypothesis: sRNA represses GFP by binding its mRNA
- Expectations:
    1. All synChiX variations should result in negative log fold change of GFP 
    2. All synChiX
- Reality
    1. (Mixed) 
        - All log fold changes of GFP are negative for chi.Bio 
        - All log fold changes of GFP are negative for plate reader EXCEPT Mutant 16 and 18
- Conclusions
    1. Possibilities for Mutant 16 and 18:
        a. The mutants were too weak to bind GFP mRNA in moderate conditions. 
            - (NO) This would be surprising, as both have a predicted DG more negative than -20 kcal/mol.
        b. The increased binding to RFP caused these mutants to have no effect on GFP mRNA binding because of competition.
            - (YES) Both have a very negative log fold change in RFP in the plate reader
            - (YES) Mutant 18 has a more positive *relative* log FC even in the Chi.Bio GFP
            

## Looking at relative fold change (FC)
Given the following hypothesis: the sRNA represses GFP by binding 
- Expectations:
    1. We expect that mutants with stronger GFP binding E lead to a more negative relative log fold change for GFP.
    2. We expect that mutants with weaker GFP binding E lead to a more positive relative log fold change for GFP.

- Reality:
    1. (YES) Mutant 164 does indeed have a more negative relative log FC in Chi Bio and in the plate reader.
    2. (MIXED) 
        - In the Chi.Bios, this is true for Mutant 39 and Mutant 18, but not for Mutants 27, 60, and 16. 
        - In the plate reader, this is true for Mutants 16, 18, 60, and 39, but not for Mutant 39.

- Conclustions
    1. Stronger GFP binding makes the sRNA more effective
    2. Maybe the Chi.Bios are unreliable not just with RFP, but also with GFP fold change.
